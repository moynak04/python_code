```python
# Create an empty dictionary
student = {}

# Insertion of elements
student["name"] = "Moynak"
student["age"] = 18
student["branch"] = "ISE-AI"

print("Dictionary after insertion:")
print(student)

# Lookup of elements
print("\nLookup operations:")
print("Name:", student["name"])
print("Age:", student.get("age"))

# Deletion of elements
del student["age"]          # delete using del
student.pop("branch")       # delete using pop()

print("\nDictionary after deletion:")
print(student)
```

```python
# Read 6 subject marks into a 1-D array
marks = []

print("Enter marks for 6 subjects:")
for i in range(6):
    mark = int(input(f"Subject {i+1}: "))
    marks.append(mark)

# Bubble Sort to arrange marks in descending order
n = len(marks)
for i in range(n - 1):
    for j in range(0, n - i - 1):
        if marks[j] < marks[j + 1]:
            # Swap
            marks[j], marks[j + 1] = marks[j + 1], marks[j]

# Display the sorted marks
print("\nStudent Report (Highest to Lowest Marks):")
for i, mark in enumerate(marks, start=1):
    print(f"{i}. {mark}")
```
